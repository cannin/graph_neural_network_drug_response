```python
import pandas as pd
import numpy as np
```

```python
class_nsc = pd.read_csv('../DrugCell/data_rcellminer/class_by_nsc.csv')
df = pd.read_csv('../data/rcellminer.csv', index_col=0)[[
    'NSC', 'NAME', 'FDA_STATUS', 'PUBCHEM_ID', 'SMILES'
]]
df = df.rename(columns={'SMILES': 'SMILES_rcellminer', 'PUBCHEM_ID': 'PUBCHEM_ID_rcellminer'})
df = df.merge(class_nsc, on='NSC', how='outer')
tmp = pd.read_csv('../data/nsc_cid_smiles.csv')
tmp = tmp.rename(columns={'SMILES': 'SMILES_pubchemAPI', 'CID': 'PUBCHEM_ID_API'})
df = df.merge(tmp, on='NSC', how='outer')
```

```python
df = df[['NSC', 'NAME', 'PUBCHEM_ID_rcellminer', 'PUBCHEM_ID_API', 'SMILES_rcellminer', 'SMILES_pubchemAPI']]
df.to_csv('./pubchem_smiles_nsc.csv')
```
