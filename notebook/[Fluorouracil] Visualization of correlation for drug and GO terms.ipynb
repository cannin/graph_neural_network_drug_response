# Fluorouracil [PubChem ID 3385](https://pubchem.ncbi.nlm.nih.gov/compound/3385)

Fluorouracil is a pyrimidine analog used to treat basal cell carcinomas, and as an injection in palliative cancer treatment.

```python
corr = pd.read_csv("../DrugCell/data_rcellminer/weight_corr.csv", index_col=0)
corr['type'] = ['GO' if 'GO:' in i else 'gene' for i in corr.index]
corr = corr.T.sort_index(ascending=False).T
```

```python
df = pd.DataFrame(corr[corr['type'] == 'GO'][SMILES]).sort_values(SMILES)
df = df[df[SMILES] != 0]
```

```python
plt.bar(
    df.index,
    df[SMILES]
)
plt.xticks(rotation=45)
```

```python
plt.bar(
    df.iloc[:20].index,
    df.iloc[:20][SMILES]
)
plt.xticks(rotation=45)
```

```python
plt.bar(
    df.iloc[-20:].index,
    df.iloc[-20:][SMILES]
)
plt.xticks(rotation=45)
```
