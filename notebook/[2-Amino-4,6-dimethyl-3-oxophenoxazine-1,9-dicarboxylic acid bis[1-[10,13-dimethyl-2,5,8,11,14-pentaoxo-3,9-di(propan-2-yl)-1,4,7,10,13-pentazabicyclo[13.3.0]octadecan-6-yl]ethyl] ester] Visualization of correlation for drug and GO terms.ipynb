```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pubchem_id =  pd.read_csv('../data/nsc_cid_smiles.csv')
translater = {pubchem_id['CID'][i]:pubchem_id['SMILES'][i] for i in pubchem_id.index}
```

```python
cid = 45006140
SMILES = translater[cid]
```

# 2-Amino-4,6-dimethyl-3-oxophenoxazine-1,9-dicarboxylic acid bis[1-[10,13-dimethyl-2,5,8,11,14-pentaoxo-3,9-di(propan-2-yl)-1,4,7,10,13-pentazabicyclo[13.3.0]octadecan-6-yl]ethyl] ester [PubChem ID 45006140](https://pubchem.ncbi.nlm.nih.gov/compound/45006140)

```python
corr = pd.read_csv("../DrugCell/data_rcellminer/weight_corr.csv", index_col=0)
corr['type'] = ['GO' if 'GO:' in i else 'gene' for i in corr.index]
corr = corr.T.sort_index(ascending=False).T
```

```python
df = pd.DataFrame(corr[corr['type'] == 'GO'][SMILES]).sort_values(SMILES)
df = df[df[SMILES] != 0]
```

```python
plt.bar(
    df.index,
    df[SMILES]
)
plt.xticks(rotation=45)
```

```python
plt.bar(
    df.iloc[:20].index,
    df.iloc[:20][SMILES]
)
plt.xticks(rotation=45)
```

```python
plt.bar(
    df.iloc[-20:].index,
    df.iloc[-20:][SMILES]
)
plt.xticks(rotation=45)
```
