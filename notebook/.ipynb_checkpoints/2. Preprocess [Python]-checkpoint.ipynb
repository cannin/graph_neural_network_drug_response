```python
import pandas as pd
import numpy as np
```

```python
drugAnnot = pd.read_csv('data/smiles_by_nsc.csv', index_col=0).dropna()
drugAnnot.head()
```

```python
nci60Act = pd.read_csv('data/nci60Act.csv', index_col=0)
nci60Act.head()
```

```python
base = pd.DataFrame()
for i in nci60Act.columns:
    tmp = pd.DataFrame((nci60Act[i].dropna()))
    tmp.columns = ['Score']
    tmp['Cell Line'] = [i]*len(tmp)
    tmp = pd.merge(tmp, drugAnnot, left_index=True, right_index=True)
    base = pd.concat([base, tmp])
    
base = base[['Cell Line', 'SMILES', 'Score']]
base = base.reset_index(drop=True)
```
